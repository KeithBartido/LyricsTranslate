# 1. NXN Numpy matrix

```python
def matrix(n, a, b):
    m = np.random.randint(a, b+1, size = (n,n))
    print(m)
```

# 2. Count plot (Top 20 occuring words)

```python
file = open("data/networks.txt", "r").read()
print(file)
```

```python
stop = open('data/stop.txt', 'r').read()
print(stop)
```

```python
filterr = []
with open('data/stop.txt', 'r')as file:
    for line in file:
        line = line.strip()
        line = line.lower()
        line = line.translate(line.maketrans("", "", string.punctuation))
        words = line.split(" ")
        filterr.append(line)
filterr
```

```python
dictionary = {}

filtered = []
with open('data/networks.txt', 'r') as file:
    for line in file:
        line = line.strip()
        line = line.lower()
        line = line.translate(line.maketrans("", "", string.punctuation))
        words = line.split(" ")
        
    for w in words:
        if w not in filterr:
            filtered.append(w)
    
    for word in filtered:
        if word in dictionary:
            dictionary[word] = dictionary[word] + 1
        else:
            dictionary[word] = 1

            
for key in list(dictionary.keys()):
    print(key, ":", dictionary[key])
```

```python
dfFiltered = " "
data = dfFiltered.join(filtered)
data
```

# Unsorted

```python
d = Counter(data.split()).items()
df = pd.DataFrame(d, columns=['word', 'count'])
df
```

# Sorted

```python
dfsorted = df.sort_values(by=['count'], ascending=False)
dfsorted = dfsorted.reset_index(drop=True)
dfsorted
```

```python
top20 = dfsorted.head(20)
top20
```

```python
wordcount = top20

%matplotlib inline
plt.figure(figsize=(13,11))
sns.set(style="darkgrid")

ax = sns.barplot(y='word', x='count', data=wordcount.reset_index(), color="salmon")
ax.set(xlabel='Word Count', ylabel='Occuring Words')
labels = dfsorted['count'].head(20)
for i, v in enumerate(labels):
    ax.text(v + 0.7, i + .10, str(v), color='black', fontweight='roman')
ax.locator_params(axis='count', integer=True)
plt.title("Top 20 Most Common Occuring Words in network.txt")
plt.show()
```

Filtering the network.txt file with the stop.txt stopword file changes the distribution of the words a bit. We're  not using the nltk stopwords in this case so "to", a common stopword was retained and sits at the top position with 56 occurences. 

Every other word after this has way fewer occurences

# 3. Convert Filipino phrases into English

```python
translator = {}
file = open('data/tl-en.txt', 'r', encoding="utf8")
for line in file:
    (eng,span) = line.split()
    translator[eng] = span    
```

```python
ogAteneo = open('data/Ateneo.txt', 'r').read()
print(ogAteneo)
```

```python
sentence = []
with open('data/Ateneo.txt', 'r') as f:
    for line in f:
        line = line.strip()
        line = line.lower()
        line = line.translate(line.maketrans("", "", string.punctuation))
        words = line.split(" ")
        for word in words:
            for key, value in translator.items():
                if word in key:
                    word = word.replace(key,value)
            sentence.append(word)
            
translation = " ".join(sentence)
print("Original text: \n")
print(ogAteneo +"\n")
print("Translated text: \n")
print(translation)
```

# 4. Convert English song to Filipino

```python
page = requests.get("https://www.azlyrics.com/lyrics/twodoorcinemaclub/undercovermartyn.html") #fetches lyrics from online
soup = BeautifulSoup(page.content, 'html.parser') 
output = soup.find_all('div')[19].get_text()
output.strip()

bestSong = output.translate(str.maketrans('', '', string.punctuation)) #removes punctuation marks for EZ translations
bestSong = bestSong.lower()
```

```python
sentence = []
undercoverMartyn = io.StringIO(bestSong) #i think it allows me to treat the string-turned html as files
with undercoverMartyn as file:  #thanks io
    for line in file:
        line = line.strip() #remove leading spaces
        words = line.split(" ") #turn that shit into a list baby
        for word in words: 
            for key, value in translator.items(): #check each key and value in the dictionary
                if word.endswith('s'): #check for pluralities
                    if word[:-1] in value: #-1 to get rid of the s
                        word = word[:-1].replace(value,key)
                        word = "mga " + word #gusto ko nang sumabog at sabihin ang masasamang *mga words*
                else:
                    if word in value:
                        word = word.replace(value, key)
                    
            sentence.append(word)
    
translation = " ".join(sentence).strip()
print("Original text: ")
print(bestSong +"\n")
print("Translated text: \n")
print(translation)
```
